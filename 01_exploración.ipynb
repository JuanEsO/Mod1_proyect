```python
df = pd.read_csv("train.csv")
```

```python
# Cantidad de estudiantes por clase
estudiantes_por_clase = df['ESTU_PRGM_ACADEMICO'].value_counts()

# Cantidad de estudiantes por departamento
estudiantes_por_departamento = df['ESTU_PRGM_DEPARTAMENTO'].value_counts()

# Cantidad de estudiantes por estrato de vivienda
estudiantes_por_estrato = df['FAMI_ESTRATOVIVIENDA'].value_counts()

# Proporción de estudiantes que tienen acceso a internet en casa
proporcion_internet = df['FAMI_TIENEINTERNET'].value_counts(normalize=True) * 100

```

```python
print("Cantidad de estudiantes por clase:\n", estudiantes_por_clase)

# Mostrar los resultados en gráficos
plt.figure(figsize=(12, 6))

# Gráfico de barras para la cantidad de estudiantes por estrato de vivienda
plt.subplot(2, 2, 3)
estudiantes_por_estrato.plot(kind='bar', color='lightgreen')
plt.title('Cantidad de Estudiantes por Estrato de Vivienda')
plt.xlabel('Estrato de Vivienda')
plt.ylabel('Cantidad de Estudiantes')

# Gráfico de pastel para la proporción de estudiantes que tienen acceso a internet en casa
plt.subplot(2, 2, 4)
proporcion_internet.plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightcoral'])
plt.title('Proporción de Estudiantes con Acceso a Internet en Casa')
plt.ylabel('')

# Gráfico de barras para la cantidad de estudiantes por departamento
plt.subplot(2, 2, 2)
estudiantes_por_departamento.plot(kind='bar', color='salmon')
plt.title('Cantidad de Estudiantes por Departamento')
plt.xlabel('Departamento')
plt.ylabel('Cantidad de Estudiantes')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
