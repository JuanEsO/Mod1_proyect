Procesar F_EDUCACIONMADRE y F_EDUCACIONPADRE

```python
df = df.copy()
df.F_EDUCACIONMADRE.values[df.F_EDUCACIONMADRE.isna()] = 'no info'
df.F_EDUCACIONMADRE = ['no info' if i in ['No sabe', 'No Aplica'] else i for i in df.F_EDUCACIONMADRE.values]
```

```python
df = df.copy()
df.F_EDUCACIONPADRE.values[df.F_EDUCACIONPADRE.isna()] = 'no info'
df.F_EDUCACIONPADRE = ['no info' if i in ['No sabe', 'No Aplica'] else i for i in df.F_EDUCACIONPADRE.values]
```

```python
df.F_EDUCACIONMADRE.value_counts()
```

```python
x = df.F_EDUCACIONMADRE.values
# x = x.astype(str)
F_EDUCACIONMADRE_vals = sorted(np.unique(x))
# F_EDUCACIONMADRE_vals.remove('nan')
F_EDUCACIONMADRE_onehot_vals = {val: np.eye(len(F_EDUCACIONMADRE_vals))[i] for i,val in enumerate(F_EDUCACIONMADRE_vals)}
```

```python
y = df.F_EDUCACIONPADRE.values
# y = y.astype(str)
F_EDUCACIONPADRE_vals = sorted(np.unique(y))
# F_EDUCACIONPADRE_vals.remove('nan')
F_EDUCACIONPADRE_onehot_vals = {val: np.eye(len(F_EDUCACIONPADRE_vals))[i] for i,val in enumerate(F_EDUCACIONPADRE_vals)}
F_EDUCACIONPADRE_onehot_vals
```

```python
F_EDUCACIONMADRE_onehot_vals
```

```python
F_EDUCACIONMADRE_onehot_enc = np.r_[[F_EDUCACIONMADRE_onehot_vals[i] for i in df.F_EDUCACIONMADRE]]
F_EDUCACIONPADRE_onehot_enc = np.r_[[F_EDUCACIONPADRE_onehot_vals[i] for i in df.F_EDUCACIONPADRE]]

F_EDUCACIONMADRE_df = pd.DataFrame(F_EDUCACIONMADRE_onehot_enc, columns=[f"F_EDUCACIONMADRE__{v}" for v in F_EDUCACIONMADRE_onehot_vals])
F_EDUCACIONPADRE_df = pd.DataFrame(F_EDUCACIONPADRE_onehot_enc, columns=[f"F_EDUCACIONPADRE__{v}" for v in F_EDUCACIONPADRE_onehot_vals])
```

```python
df = pd.concat([F_EDUCACIONMADRE_df, df], axis=1).drop('F_EDUCACIONMADRE', axis=1)
df = pd.concat([F_EDUCACIONPADRE_df, df], axis=1).drop('F_EDUCACIONPADRE', axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le_academico = LabelEncoder()
le_departamento = LabelEncoder()
```

```python
df.E_PRGM_ACADEMICO.values[df.E_PRGM_ACADEMICO.isna()] = 'no info'
df.E_PRGM_DEPARTAMENTO.values[df.E_PRGM_DEPARTAMENTO.isna()] = 'no info'
```

```python
df['E_PRGM_ACADEMICO'] = le_academico.fit_transform(df['E_PRGM_ACADEMICO'])
df['E_PRGM_DEPARTAMENTO'] = le_departamento.fit_transform(df['E_PRGM_DEPARTAMENTO'])
```
