```python
import pandas as pd
import numpy as np

df = pd.read_csv("train.csv")
```

Eliminar las columnas ID

```python
df = df.drop(columns=['ID'])
```

Procesar columna ESTU_VALORMATRICULAUNIVERSIDAD

```python
df.ESTU_VALORMATRICULAUNIVERSIDAD.values[df.ESTU_VALORMATRICULAUNIVERSIDAD.isna()] = 'no info'
```

```python
cmap = {'Entre 1 millón y menos de 2.5 millones': 1.75,
 'Entre 2.5 millones y menos de 4 millones': 3.25,
 'Menos de 500 mil': .250,
 'Entre 500 mil y menos de 1 millón': .75,
 'Entre 4 millones y menos de 5.5 millones': 4.75,
 'Más de 7 millones': 7.75,
 'Entre 5.5 millones y menos de 7 millones': 6.25,
 'No pagó matrícula': 0,
 'no info': -1}

df.ESTU_VALORMATRICULAUNIVERSIDAD = np.r_[[cmap[i] for i in df.ESTU_VALORMATRICULAUNIVERSIDAD]]
df.ESTU_VALORMATRICULAUNIVERSIDAD.value_counts()
```

Procesar FAMI_EDUCACIONMADRE y FAMI_EDUCACIONPADRE

```python
df = df.copy()
df.FAMI_EDUCACIONMADRE.values[df.FAMI_EDUCACIONMADRE.isna()] = 'no info'
df.FAMI_EDUCACIONMADRE = ['no info' if i in ['No sabe', 'No Aplica'] else i for i in df.FAMI_EDUCACIONMADRE.values]
```

```python
df = df.copy()
df.FAMI_EDUCACIONPADRE.values[df.FAMI_EDUCACIONPADRE.isna()] = 'no info'
df.FAMI_EDUCACIONPADRE = ['no info' if i in ['No sabe', 'No Aplica'] else i for i in df.FAMI_EDUCACIONPADRE.values]
```

```python
df.FAMI_EDUCACIONMADRE.value_counts()
```

```python
x = df.FAMI_EDUCACIONMADRE.values
# x = x.astype(str)
FAMI_EDUCACIONMADRE_vals = sorted(np.unique(x))
# FAMI_EDUCACIONMADRE_vals.remove('nan')
FAMI_EDUCACIONMADRE_onehot_vals = {val: np.eye(len(FAMI_EDUCACIONMADRE_vals))[i] for i,val in enumerate(FAMI_EDUCACIONMADRE_vals)}
```

```python
y = df.FAMI_EDUCACIONPADRE.values
# y = y.astype(str)
FAMI_EDUCACIONPADRE_vals = sorted(np.unique(y))
# FAMI_EDUCACIONPADRE_vals.remove('nan')
FAMI_EDUCACIONPADRE_onehot_vals = {val: np.eye(len(FAMI_EDUCACIONPADRE_vals))[i] for i,val in enumerate(FAMI_EDUCACIONPADRE_vals)}
FAMI_EDUCACIONPADRE_onehot_vals
```

```python
FAMI_EDUCACIONMADRE_onehot_vals
```

```python
FAMI_EDUCACIONMADRE_onehot_enc = np.r_[[FAMI_EDUCACIONMADRE_onehot_vals[i] for i in df.FAMI_EDUCACIONMADRE]]
FAMI_EDUCACIONPADRE_onehot_enc = np.r_[[FAMI_EDUCACIONPADRE_onehot_vals[i] for i in df.FAMI_EDUCACIONPADRE]]

FAMI_EDUCACIONMADRE_df = pd.DataFrame(FAMI_EDUCACIONMADRE_onehot_enc, columns=[f"FAMI_EDUCACIONMADRE__{v}" for v in FAMI_EDUCACIONMADRE_onehot_vals])
FAMI_EDUCACIONPADRE_df = pd.DataFrame(FAMI_EDUCACIONPADRE_onehot_enc, columns=[f"FAMI_EDUCACIONPADRE__{v}" for v in FAMI_EDUCACIONPADRE_onehot_vals])
```

```python
df = pd.concat([FAMI_EDUCACIONMADRE_df, df], axis=1).drop('FAMI_EDUCACIONMADRE', axis=1)
df = pd.concat([FAMI_EDUCACIONPADRE_df, df], axis=1).drop('FAMI_EDUCACIONPADRE', axis=1)
```
